```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
print(x)
```

```python
y=2*x
```

```python
y
```

# Line chart

```python
plt.plot(x,y)
plt.title("line chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# Bar chart

```python
plt.bar(x,y)
plt.title("Bar chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
